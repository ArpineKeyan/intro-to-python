## Numpy Overview

NumPy, short for Numerical Python, is one of the most important foundational packages for numerical computing in Python. Most computational packages providing
scientific functionality use NumPy’s array objects as the lingua franca for data
exchange.

Array-oriented computing in Python traces its roots back to 1995,
when Jim Hugunin created the Numeric library. Over the next 10
years, many scientific programming communities began doing
array programming in Python, but the library ecosystem had
become fragmented in the early 2000s. In 2005, Travis Oliphant
was able to forge the NumPy project from the then Numeric and
Numarray projects to bring the community together around a single array computing framework.

### Boolean Matrices

```
a = np.array([1, 2, 3])
b = np.array([1, 2, 0])

a == b, a > b, a >= b, a < b, a <= b
```

```
a = np.array([True, False, True])
b = np.array([False, False, True])

# logical elementwise operations
print(f'a and b {a & b}')
print(f'a or b {a | b}')
print(f'not a {~a}')
print(f'a xor b {a ^ b}')
```

```
a.any(), a.all() # logical operations on all elements 
```

```
# slice with conditions 
a = np.array([np.nan, 1,2,np.nan,3,4,5]) 
print(a)
print(a[~np.isnan(a)])
~np.isnan(a)
```

```
a = np.array([0, 1, 2, 3,4,5,6])
print(a)
print(a[[True, False, True, True, True, True, False]])
print(a[a > 1])
```

## Linear Algebra 

* Dot product $a~\cdot~b = (a_1, a_2, .., a_n) \cdot (b_1, b_2, .., b_n) = a_1b_1 + a_2b_2 + .. + b_nb_n = \sum_{i=1}^{n} a_ib_i$:


```
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
print(a,b)
```

```
np.dot(a, b)
```

```
a.dot(b)
```

```
np.matmul(a, b)  #matrix multiplication
```

```
a @ b # calls matmul
```

```
a = np.array([[1,2],[3,4]]) 
b = np.array([[11,12],[13,14]]) 
np.dot(a,b)
```

```
np.matmul(a,b)
```

``np.matmul`` and ``np.dot`` differn in operation with tensors, but we won't dive in

```
a = np.array([[1, 2], [3, 4]])
b = np.transpose(a)
c = a.T
```

```
print("Matrix\n", a)
print("Transposed with function:\n", b)
print("Transposed with method:\n",  c)
```

```
from numpy import linalg # more linear algebry functions
```

```
A = np.array([[1, 1], [1, 0]])
x = np.array([[4, 1], [2, 2]])
b = np.dot(A, x)
print(b)
```

* Solve linear system __$Ax=b$__:

```
x = np.linalg.solve(A, b)
print(x)
```
